```python
ALL = "All"
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique
```

```python
dropdown_subject = widgets.Dropdown(options = unique_sorted_values_plus_ALL(df1.columns))
```

```python
def dropdown_subject_handler(change):
    if (change.new == ALL):
        display(df1)
    else:
        display(df1[df1.columns == change.new])
```

```python
dropdown_subject.observe(dropdown_subject_handler, names = "value")
```
